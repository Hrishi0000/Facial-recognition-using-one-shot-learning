```python

# OpenCV loads images with color channels
# in BGR order. So we need to reverse them
def load_image(path):
    img = cv2.imread(path, 1)
    return img[...,::-1]

# Initialize the OpenFace face alignment utility
alignment = AlignDlib('shape_predictor_68_face_landmarks.dat')

def show_original_and_aligned(path):
    plt.figure(figsize=(10,10))
    # Load an image
    original = load_image(path)

    # Detect face and return bounding box
    bb = alignment.getLargestFaceBoundingBox(original)

    # Transform image using specified face landmark indices and crop image to 96x96
    aligned = alignment.align(96, original, bb, landmarkIndices=AlignDlib.OUTER_EYES_AND_NOSE)

    # Show original image
    plt.subplot(131)
    plt.imshow(original)

    # Show original image with bounding box
    plt.subplot(132)
    plt.imshow(original)
    plt.gca().add_patch(patches.Rectangle((bb.left(), bb.top()), bb.width(), bb.height(), fill=False, color='red'))

    # Show aligned image
    plt.subplot(133)
    plt.imshow(aligned)
```

```python
show_original_and_aligned('images/3.jpg')
```

```python
# def align_image(img):
#     '''
#     Objective = Get all aligned faces 
#     Parameters:
#     img = RGB image in ndarray format
#     return - list of faces in ndarray format
#     '''
#     aligned=[]
#     res = get_bb(img)
#     for bb in res:
#         aligned.append(alignment.align(96, img, bb, landmarkIndices=AlignDlib.OUTER_EYES_AND_NOSE))
#     return aligned

def align_image(img):
    return alignment.align(96, img, alignment.getLargestFaceBoundingBox(img), landmarkIndices=AlignDlib.OUTER_EYES_AND_NOSE)
```

```python
# embedded = np.zeros((1, 128))

# def generate_encoding(img):
#     '''
#     Objective - generate encoding based on FACE model
#     parameter:
#     img = ALIGNED RGB face image in ndarray format
#     return - 128 dimensional encoding
#     '''   
    
#     # scale RGB values to interval [0,1]
#     img = (img / 255.).astype(np.float32)
#     # obtain embedding vector for image
#     embedded= nn4_small2.predict(np.expand_dims(img, axis=0))[0]
#     return embedded

embedded = np.zeros((1, 128))

def generate_encoding(path):
  #print(m.image_path())    
  img = load_image(path)
  img = align_image(img)
  # plt.imshow(img)
  # scale RGB values to interval [0,1]
  img = (img / 255.).astype(np.float32)
  # obtain embedding vector for image
  embedded= nn4_small2.predict(np.expand_dims(img, axis=0))[0]
  return embedded
```

```python
def distance(emb1, emb2):
    return np.sum(np.square(emb1 - emb2))
```

```python
database={}
```

```python


# def make_database(entries):
#     for (i,j) in entries.items():
#         if i not in database.keys():
#             img = load_image(entries[i])
#             img = align_image(img)
#             database[i] = generate_encoding(img[0])

def make_database(entries):
  for (i,j) in entries.items():
    if i not in database.keys():
      database[i] = generate_encoding(j)
    
```

```python
# id=['hrishi','shaima','debanshi']
# photos=['/content/hrishi1.jpg','/content/shaima2.jpg','debanshi1.jpg']
# make_database(id,photos)

entries={}
entries['Hrishi']='images/hrishi1.jpg'
entries['Shaima']='images/shaima3.jpg'
entries['Debanshi']='images/debanshi2.jpg'
make_database(entries)
```

```python
threshold = 0.7

def verify(id,path):
    '''
    OBJECTIVE - FACE VERIFICATION
    Parameters:
    id- String, Person name as in database to be verified
    path - Path to a SINGLE image
    Return - String
    '''
    img = load_image(path)
    img = align_image(img)
    anchor = database[id]
    test_em = generate_encoding(img[0])
    if distance(anchor,test_em) < threshold:
        print(distance(anchor,test_em) )
        print("Yes it  is ",id)
    else:
        print("Verification Failed")
```

```python
import cv2

cam = cv2.VideoCapture(0)

cv2.namedWindow("test")

img_counter = 0

while True:
    ret, frame = cam.read()
    if not ret:
        print("failed to grab frame")
        break
    cv2.imshow("test", frame)

    k = cv2.waitKey(1)
    if k%256 == 27:
        # ESC pressed
        print("Escape hit, closing...")
        break
    elif k%256 == 32:
        # SPACE pressed
        img_name = "opencv_frame_{}.png".format(img_counter)
        cv2.imwrite(img_name, frame)
        print("{} written!".format(img_name))
        img_counter += 1

cam.release()

cv2.destroyAllWindows()
```

```python
def get_bb(img):
    '''
    Objective = Return Bounding Box(s)
    parameter:
    img = RGB image in ndarray format
    return = list of dlib.rectangle objects
    '''
    alignment = AlignDlib('shape_predictor_68_face_landmarks.dat')
    bb = alignment.getAllFaceBoundingBoxes(img)
    return bb
```

```python
img = load_image('images/3.jpg')
boxes = get_bb(img)
```
